# RECORDS AEP 2025

```python
import pandas as pd

records_25= pd.read_csv('records_aep25.csv', sep= ';', decimal=',')

records_25
```

```python
records_25.drop(columns= ['Unnamed: 5'], inplace=True)

records_25.columns
```

### LIMPIEZA DE DATOS

```python
records_25['NOMBRE']= records_25['NOMBRE'].replace('Pablo Sánchez Urriza', 'Sánchez Urriza Pablo')
records_25['NOMBRE']= records_25['NOMBRE'].replace({
    'Álvaro Fernández Arcas': 'Fernández Arcas Álvaro',
    'Enrique Sánchez Quintero': 'Sánchez Quintero Enrique',
    'Inmaculada Soto Ruíz': 'Soto Ruíz Inmaculada',
    'Ashkan Aghasafari Farhangi': 'Aghasafari Farhangi Ashkan',
    'Emma Hassan Trujillo': 'Hassan Trujillo Emma'
})
```

```python
for v in records_25["club"].unique():
    if "INTEND POWER" in v:
        print(repr(v))
```

```python
records_25['club']= records_25['club'].replace('INTEND POWER MADRID', 'INTEND POWER')
```

## EXPLORACIÓN DE DATOS

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
porcentaje_genero = records_25['sexo'].value_counts(normalize=True) * 100
porcentaje_genero.plot(kind='bar', color=['pink', 'blue'])
plt.title('Porcentaje de Participación por Género')
plt.xlabel('Género')
plt.ylabel('Porcentaje (%)')
plt.text(0, 55, f'{porcentaje_genero["F"]:.1f}%', ha='center', va='bottom', color='black', fontsize=12)
plt.text(1, 45, f'{porcentaje_genero["M"]:.1f}%', ha='center', va='bottom', color='black', fontsize=12)
plt.xticks(rotation=0)
plt.show()
```

```python
atletas= records_25['NOMBRE'].drop_duplicates()
atletas.nunique()
```

```python
porcentaje_atletas= (records_25['cat_edad'].value_counts(normalize=True) * 100).round(2)

porcentaje_atletas.plot(kind='bar')
plt.xlabel('Categoría de Edad')
plt.ylabel('Porcentaje de Atletas')
plt.text(0, 25, f'{porcentaje_atletas.iloc[0]}%', ha='center', va='bottom')
plt.text(1, 20, f'{porcentaje_atletas.iloc[1]}%', ha='center', va='bottom')
plt.text(2, 15, f'{porcentaje_atletas.iloc[2]}%', ha='center', va='bottom')
plt.text(3, 10, f'{porcentaje_atletas.iloc[3]}%', ha='center', va='bottom')
plt.text(4, 10, f'{porcentaje_atletas.iloc[4]}%', ha='center', va='bottom')
plt.text(5, 10, f'{porcentaje_atletas.iloc[5]}%', ha='center', va='bottom')
plt.text(6, 10, f'{porcentaje_atletas.iloc[6]}%', ha='center', va='bottom')
plt.title('Distribución de Atletas por Categoría de Edad en AEP 2025')
plt.show()
```

```python
porcentaje_peso= (records_25['cat_peso'].value_counts(normalize=True) * 100).round(1)
ax= porcentaje_peso.plot(kind='bar')
plt.xlabel('Categoría de Peso')
plt.ylabel('Porcentaje de Atletas')
plt.title('Porcentaje de categorías de peso en AEP 2025')

for i, valor in enumerate(porcentaje_peso):
    ax.text(i, valor + 0.5, f'{valor}', ha='center', va='bottom')

plt.show()
```

```python
ranking_atletas_masc = (
    records_25[records_25['sexo']== 'M']
    .drop_duplicates(
        subset=["NOMBRE", "mov"]
    )
    .groupby("NOMBRE")
    .size()
    .reset_index(name="n_records")
    .sort_values("n_records", ascending=False)
)
ranking_atletas_masc
```

```python
ranking_atletas_fem = (
    records_25[records_25['sexo']== 'F']
    .drop_duplicates(
        subset=["NOMBRE", "mov"]
    )
    .groupby("NOMBRE")
    .size()
    .reset_index(name="n_records")
    .sort_values("n_records", ascending=False)
)
ranking_atletas_fem
```

### COMPETIONES Y CLUBES

```python
comp_levels = (
    records_25[['NOMBRE', 'nivel']]
    .drop_duplicates()
)
comp_count = (
    comp_levels
    .value_counts('nivel')
    .sort_index()
)
plt.figure()
comp_count.plot(kind='bar', color='red')
plt.xlabel('Nivel competitivo')
plt.ylabel('Número de competiciones')
plt.title('Número de competiciones por nivel competitivo (2025)')
plt.tight_layout()
plt.show()
```

### RECORDS DE CADA MOVIMIENTO

```python
(records_25['mov'].value_counts(normalize=True) *100).round(2)
```

```python
ranking_club_mov = (
    records_25
    .drop_duplicates(
        subset=["club", "NOMBRE", "COMPETICIÓN", "mov"]
    )
    .groupby(["mov", "club"])
    .size()
    .reset_index(name="n_records")
    .sort_values(["mov", "n_records"], ascending=[True, False])
)

tabla_medallero = (
    ranking_club_mov
    .pivot(index="club", columns="mov", values="n_records")
    .fillna(0)
    .astype(int)
    .reset_index()
)

cols_mov = tabla_medallero.columns.difference(["club"])

tabla_medallero["Nº de Récords Totales"] = tabla_medallero[cols_mov].sum(axis=1)

tabla_medallero.sort_values(by= 'Nº de Récords Totales', ascending=False)
```
